```python
import json
import pandas as pd
import numpy as np
```

```python
wash_conc_file = "D:\\projects\\nmf_py\\user_data\\wash_con_1.csv"
wash_unc_file = "D:\\projects\\nmf_py\\user_data\\wash_unc_1.csv"
```

```python
wash_conc_df = pd.read_csv(wash_conc_file)
wash_conc_labels = [col for col in wash_conc_df.columns if "val" in col]
wash_conc_labels.insert(0, "obs_date")
wash_conc_df = wash_conc_df[wash_conc_labels]
wash_conc_df["obs_date"] = pd.to_datetime(wash_conc_df["obs_date"], format="%Y%m%d")
wash_conc_df.set_index("obs_date", inplace=True)
wash_conc_df
```

```python
wash_unc_df = pd.read_csv(wash_unc_file)
wash_unc_labels = [col for col in wash_unc_df.columns if "unc" in col]
wash_unc_labels.insert(0, "obs_date")
wash_unc_df = wash_unc_df[wash_unc_labels]
wash_unc_df["obs_date"] = pd.to_datetime(wash_unc_df["obs_date"], format="%Y%m%d")
wash_unc_df.set_index("obs_date", inplace=True)
wash_unc_df
```

```python
val_medians = wash_conc_df.median()
val_medians
```

```python
drop_columns = {k: v for k, v in val_medians.items() if v == -999.0}
drop_columns
```

```python
wash_conc_df.drop(labels=drop_columns.keys(), inplace=True, axis=1)
wash_conc_df
```

```python
unc_drop_columns = [col.replace("val", "unc") for col in drop_columns.keys()]
wash_unc_df.drop(labels=unc_drop_columns, inplace=True, axis=1)
wash_unc_df
```

```python
conc_columns = {}
for col in wash_conc_df.columns:
    conc_columns[col] = col.split("_val")[0]
wash_conc_df.rename(columns=conc_columns, inplace=True)
```

```python
wash_conc_df.replace(-999.0, np.nan, inplace=True)
wash_conc_df.fillna(wash_conc_df.median(), inplace=True)
wash_conc_df[wash_conc_df == 0.0] = 1e-15
wash_conc_df
```

```python
unc_columns = {}
for col in wash_unc_df.columns:
    unc_columns[col] = col.split("_unc")[0]
wash_unc_df.rename(columns=unc_columns, inplace=True)
```

```python
wash_unc_df.replace(-999.0, np.nan, inplace=True)
unc_mod = wash_conc_df * 0.06
wash_unc_df.update(unc_mod)
wash_unc_df = wash_unc_df.abs() + 1e-15
wash_unc_df
```

```python
cleaned_wash_conc_file = "D:\\projects\\nmf_py\\user_data\\wash_con_cleaned.csv"
cleaned_wash_unc_file = "D:\\projects\\nmf_py\\user_data\\wash_unc_cleaned.csv"
wash_conc_df.to_csv(cleaned_wash_conc_file)
wash_unc_df.to_csv(cleaned_wash_unc_file)
```
